#### Download MOD11C3 data 2000-2025 ([Publication URL lpdaac](https://opendap.cr.usgs.gov/opendap/hyrax/DP137/MOLT/MOD11C3.061/contents.html))

```python
# download data
import requests

# Read URLs from the file
with open("MOD11C3/urls.txt", "r") as file:
    urls = file.readlines()

# Download each file
for url in urls:
    url = url.strip()  # Remove any extra whitespace
    response = requests.get(url)
    filename = url.split("/")[-1]
    with open(filename, "wb") as f:
        f.write(response.content)
```

#### Daytime Land Surface Temperature

```python
import numpy as np
import matplotlib.pyplot as plt
import netCDF4

# Open the NetCDF file
file_name = 'MOD11C3.A2025032.061.2025063170308.hdf'
var = 'LST_Day_CMG'

with netCDF4.Dataset(file_name, 'r') as dataset:
    variable = dataset.variables[var]
    
    # Extract data
    data = variable[:].flatten()  # Flatten for histogram

    # Get the length of the variable
    #length = len(data)
    #print(f"Length of '{var}': {length}")

    # Handle missing values (_FillValue)
    fill_value = getattr(variable, '_FillValue', np.nan)
    data = np.where(data == fill_value, np.nan, data)

    # Retrieve scale factor from attributes
    scale_factor = getattr(variable, 'scale_factor', 1)  # Default to 1 if missing

    # Apply scale factor
    #data = data * scale_factor  

    # Remove NaNs for plotting
    data = data[~np.isnan(data)]

    # Plot histogram
    plt.figure(figsize=(10, 5))
    plt.hist(data, bins=50, color='blue', alpha=0.7, edgecolor='black')
    plt.xlabel("Daytime Land Surface Temperature [K]")
    plt.ylabel("Frequency")
    #plt.title(f"Histogram of Scaled LST_Day_CMG (Scale Factor: {scale_factor})")
    plt.grid(True)
    plt.show()
```
