```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (adjust the path to where your file is located)
#df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Europe_NOFLAG.csv')
#df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Oceania_NOFLAG.csv')
#df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Africa_NOFLAG.csv')
#df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Americas_NOFLAG.csv')
df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Asia_NOFLAG.csv')

# Step 1: Filter by "Area" == 'Albania'
df_albania = df[df['Area'] == 'Syrian Arab Republic']

# Step 2: Filter by "Item" == 'Apples'
df_apples = df_albania[df_albania['Item'] == 'Rice']

# Step 3: Filter by "Element" == 'Production'
df_production = df_apples[df_apples['Element'] == 'Production']

# Step 4: Check units to see if 't' (tons) is present
df_production_tons = df_production[df_production['Unit'] == 't']

# Step 5: Select the year columns (Y1961 to Y2023)
year_columns = [col for col in df_production_tons.columns if col.startswith('Y')]

# Step 6: Extract the production data for years (Y1961 to Y2023)
df_years = df_production_tons[year_columns].iloc[0]  # Take the first row (only one row for Albania)

# Step 7: Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df_years.index, df_years.values, marker='o', color='b', linestyle='-', linewidth=2)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Production (tons)')
plt.title('Rice Production in Syrian Arab Republic (1961-2023)')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (adjust the path to where your file is located)
#df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Europe_NOFLAG.csv')
#df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Oceania_NOFLAG.csv')
#df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Africa_NOFLAG.csv')
#df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Americas_NOFLAG.csv')
df = pd.read_csv('FAOSTAT/Production_Crops_Livestock_E_Asia_NOFLAG.csv')
# Areas to loop through
areas = ['Syrian Arab Republic', 'Afghanistan']
items = ['Beans, dry', 'Cattle', 'Rice', 'Wheat', 'Meat, Total', 'Sugar Crops Primary', 'Vegetables Primary']
start_year = 2000
end_year = 2020

# Loop through each area
for area in areas:
    print(f"Analyzing data for {area}...\n")
    
    # Filter data for the current area
    df_area = df[df['Area'] == area]

    # Get all unique "Item" values for the current area
    #items = df_area['Item'].unique()

    # Loop through each "Item"
    for item in items:
        print(f"\nAnalyzing {item}...\n")
        
        # Filter data for the current item
        df_item = df_area[df_area['Item'] == item]
        
        # Get all unique "Element" values for the current item
        elements = df_item['Element'].unique()
        
        # Loop through each "Element"
        for element in elements:
            print(f"\nAnalyzing Element: {element}...\n")
            
            # Filter data for the current element
            df_element = df_item[df_item['Element'] == element]
            
            # Get all unique "Unit" values for the current element
            units = df_element['Unit'].unique()
            
            # Loop through each "Unit"
            for unit in units:
                print(f"\nAnalyzing Unit: {unit}...\n")
                
                # Filter data for the current unit
                df_unit = df_element[df_element['Unit'] == unit]
                
                # Select the year columns (Y1961 to Y2023)
                year_columns = [col for col in df_unit.columns if col.startswith('Y') and start_year <= int(col[1:]) <= end_year]
                
                if not df_unit.empty and year_columns:
                    # Extract the data for years (Y1961 to Y2023)
                    df_years = df_unit[year_columns].iloc[0]  # Take the first row for the specific combination
                    
                    # Plot the data
                    plt.figure(figsize=(10, 6))
                    plt.plot(df_years.index, df_years.values, marker='o', color='b', linestyle='-', linewidth=2)
                    
                    # Set labels and title
                    plt.xlabel('Year')
                    plt.ylabel(f'{element} ({unit})')
                    plt.title(f'{item} - {element} in {area} ({unit}) (2000-2020)')
                    
                    # Rotate the x-axis labels for better readability
                    plt.xticks(rotation=45)
                    
                    # Show the plot
                    plt.tight_layout()
                    plt.show()

                else:
                    print(f"No data available for {area}, {item}, {element}, {unit}.\n")
    
    print(f"\nFinished analyzing data for {area}.\n")
```
